```python
print(sales.shape)
print(sales[['id', 'item_id', 'dept_id', 'cat_id', 'store_id', 'state_id']].nunique())
```

```python
daily_sales = sales.drop(columns=['id','item_id','dept_id','cat_id','store_id','state_id'])
print(daily_sales)
```

```python
total_sales = daily_sales.sum(axis=0)
print(total_sales)
```

```python
date_map = calendar[['d', 'date']]
total_sales = total_sales.reset_index()
```

```python
total_sales.columns = ['d', 'sales']
total_sales = total_sales.merge(date_map, on='d', how='left')
```

```python
print(date_map)
print(total_sales)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,5))
plt.plot(pd.to_datetime(total_sales['date']), total_sales['sales'])
plt.title('Total Sales Over Time')
plt.show()
```

```python
sales_numeric = sales.drop(columns=['id', 'item_id', 'dept_id', 'cat_id', 'store_id', 'state_id'])

sales_with_cat = pd.concat([sales['cat_id'], sales_numeric], axis=1)

cat_sales = sales_with_cat.groupby('cat_id').sum().sum(axis=1).sort_values(ascending=False)

cat_sales_pct = cat_sales / cat_sales.sum() * 100

cat_sales_pct.plot(kind='bar', title='Total Sales Percentage by Category', ylabel='Percentage (%)')
```

```python

store_with_sales = pd.concat([sales['store_id'], sales_numeric], axis=1)

store_id = store_with_sales.groupby('store_id').sum().sum(axis=1).sort_values(ascending=False)

store_id_pct = store_id / store_id.sum() * 100

store_id_pct.plot(kind='bar', title='Total Sales Percentage by Store', ylabel='Percentage (%)')
```
